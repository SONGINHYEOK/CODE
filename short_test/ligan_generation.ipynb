```python
for i, file in enumerate(file_list):
    file_name = path + '/' + file
    sdf = Chem.SDMolSupplier(file_name)
    mol_3d_list = open(file_name, 'r').read().split('$$$$')
    for j, mol in enumerate(sdf):
        mol_3d = mol_3d_list[j]
        if mol == None:
            f = open('error_list.sdf', 'a')
            f.write(mol_3d)
            f.write('\n')
            f.close()
            print(i, j, '번째 화합물 오류')
            continue
        else:
            g = open(path + '/1V4S_liGAN_dock/'+'DEMO_' + str(i) + '_lig_gen_fit_add_gna'+'_'+ str(j)+'.sdf',  'a')
            g.write(mol_3d.strip())
            g.write('\n')
            g.close()
        
        with open(path + '/1V4S_liGAN_dock/' + 'dock' + '_input.types', 'a') as ig:
                ig.write('0 0 ' + './data/3d_gen/1V4S_liGAN_dock/1V4S_rec.pdb'+ ' ' + './data/3d_gen/1V4S_liGAN_dock/' + 'DEMO_' + str(i) + '_lig_gen_fit_add_gna'+'_'+ str(j)+'.sdf' + '\n') 
                ig.close() 
```
