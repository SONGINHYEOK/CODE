```python
#사용 라이브러리
import os
from urllib.parse import urljoin
import requests
from bs4 import BeautifulSoup, Tag
#저장경로
OUT_DIR = '/Users/song-inhyeok/Documents/coding/PDB/PDB'
#crawling url
root_url = 'https://files.rcsb.org/download/'
html = requests.get(root_url).text
soup = BeautifulSoup(html, 'lxml')

slist = []
tables = soup.find('div', {"class": "singletablecontainer"})  # type: Tag
for a_tag in tables.find_all('a'):
    href = a_tag.get('href')
    slist.append(urljoin(root_url, href))  # 상대경로를 절대경로 url로 바꿔서 append

amount = len(slist)
for i, url in enumerate(slist):  # url는 slist안의 링크들

    response = requests.get(url)  # url에 요청
    file_name = os.path.basename(url)

    if file_name:
       file_path = os.path.join(OUT_DIR, file_name)
#file save
    with open(file_path, "wb") as pdf:
        print('({}/{}) {}'.format(i + 1, amount, file_path))
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                pdf.write(chunk)
```
